```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Heart Disease data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
col = df.target.value_counts()
```

```python
col
```

```python
# plt.figure(figsize=(10,10))
plt.bar(df.target.value_counts().index, df.target.value_counts().values)
plt.xlabel('Target Values')
plt.ylabel('Frequency')
plt.title('Distribution of Target Values')
plt.show()
```

```python
# plt.figure(figsize=(10,10))
plt.bar(df.sex.value_counts().index, df.sex.value_counts().values)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Target Values')
plt.show()
```

```python
grouped_data = df.groupby(['age', 'target']).size().reset_index(name='count')
```

```python
grouped_data
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=grouped_data, x='age', y='count', hue='target')
plt.show()
```
